### Arithmetic Operations

```python
import pandas as pd
```

```python
d={"a":[1,2,3,4,5],"s":[1,2,3,4,56],1:[12,3,4,5,5]}
var=pd.DataFrame(d)
var
```

```python
var["c"]=var["a"]+var["s"]
var
```

```python
# Inserting a column (columnIndex,columnname,values)
var.insert(1,"Python",var["c"])
var
```

```python
var["Python"]=var["c"][:3]
var
```

```python
# delete a column
var.pop("c")
var
```

```python
del var["a"]
var
```
