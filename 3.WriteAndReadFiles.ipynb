### Read and write files

```python
import pandas as pd
```

```python
d={"a":[1,2,3,4,5],"s":[1,2,3,4,56],1:[12,3,4,5,5]}
var=pd.DataFrame(d)
var.to_csv("data.csv",index=False,header=["A","B","C"])
```

```python
# Read data from csv file
csv_file=pd.read_csv("./users_data.csv")
csv_file
```

```python
# Read particular column
pd.read_csv("./users_data.csv",usecols=["name","email"]) # choose column by col_name
pd.read_csv("./users_data.csv",usecols=[0,1]) # choose column by col_index
```

```python
pd.read_csv("./users_data.csv",skiprows=[1,2]) # skip ith from top
```

```python
pd.read_csv("./users_data.csv",index_col="name") # set row names to column "name"
```

```python
pd.read_csv("./users_data.csv",header=2) # set header to row i
```

```python
pd.read_csv("./users_data.csv",names=["cols1","cols2","cols3","cols4","cols5","cols6","cols7","cols8"]) # set column names from right
```

```python
pd.read_csv("./users_data.csv",header=None)
```
