### Missing Values

```python
import pandas as pd
csv=pd.read_csv("./data_with_missing_values.csv")
csv
```

```python
# deleting row(axis=0)/col(axis=1) if it has missing value
csv.dropna(axis=0) # return not modified
csv.dropna(axis=1) # return not modified
```

```python
# only that will affect which key selected
csv.dropna(subset=["purchase_amount"],inplace=True) # modify to dataframe
# thresh=N keeps rows that have at least N non-NA values.
csv.dropna(thresh=1)
```

```python
# Filling missing values
csv=pd.read_csv("./data_with_missing_values.csv")
csv.fillna({"purchase_amount":"name"})
```

```python
csv.fillna(method="bfill",axis=0) # bottom value filled
# The limit=1 ensures that only one missing value per column is filled.
csv.fillna("python",limit=1)
```

```python
# Replace values
csv.replace(to_replace=1,value=22)
csv.replace([1,2,3,4,5,6,7,8,99,100],value=22)
```

```python
csv.replace("[A-Za-z]","p",regex=True)
```
