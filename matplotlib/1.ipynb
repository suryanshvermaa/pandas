### Matplotlib

```python
import matplotlib.pyplot as plt
plt.plot([1,5,2,4])
plt.show()
```

```python
x=[1,2,3,4]
dsa_marks=[90,94,100,91]
ml_marks=[93,98,91,96]
plt.plot(x,dsa_marks,label="dsa_marks",color='r')
plt.plot(x,ml_marks,label="ml_marks",color='g')
plt.title("Marks vs no Of Study")
plt.xlabel("no of hours studied")
plt.ylabel("Marks obtained")
plt.xticks([1,2,3,4])
plt.yticks([90,91,92,93,94,95,96,97,98,99,100])
plt.legend(loc="lower right")
plt.show()
```
